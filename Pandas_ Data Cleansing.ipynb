## Impute infinite values 

```python
dict={'a':np.arange(1,4,0.5),
     'b':np.arange(5,8,0.5),
     'c':np.arange(3,6,0.5)}
dff=pd.DataFrame(dict)
dff
```

```python
## introducing infite values
dff=dff.replace([3,5,7],-np.inf)
dff
```

```python
##converting infinite values to NaN values
#method 1
dff[np.isfinite(dff)]
```

```python
# method 2
dff=dff.replace([np.inf,-np.inf],np.nan)
dff
```

## Outliers 

```python
dict_={'Names':['John', 'Paul', "Mark", "Sarah", "Morgan", "Mike", "Ram",
            "Jennifer", "Amy", "Chris", "Mitchel", "Alex"],
    'Age':[25, 45, 12, 95, 6, 21, 34, 24, 41, 31, 19, 16]}
dff=pd.DataFrame(dict_)
dff
```

```python
from scipy import stats
```

```python
stats.zscore(dff['Age'])
```

```python
dff=dff[abs(stats.zscore(dff['Age'])) < 1.96]
dff=dff.reset_index(drop=True)
dff
```
