```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('titanic_train.csv')
df.head()
```

## Piechart 

```python
dff=df['Pclass'].value_counts().sort_index()
dff
```

```python
colors=['#191970', '#001CF0', '#0038E2']
explode=(0,0.01,0.02)
```

```python
dff.plot(kind='pie',colors=colors, explode=explode, autopct='%1.1f%%', fontsize=12, figsize=(10, 10));
```

## Line plot

```python
df['Fare'].plot(figsize=(12,5),color='Red')
plt.xlim(0,500)
plt.ylim(0,700)
plt.xlabel('index')
plt.ylabel('Fare');
```

## Scatter Plot 

```python
df.plot.scatter('Age','Fare',figsize=(12,5))
```

## Box Plot 

```python
df['Age'].plot.box(figsize=(8,5));
```

```python
df.boxplot(figsize=(12,5));
```

## Histogram 

```python
df['Age'].hist(bins=30,figsize=(12,5));
plt.xlim(0,90)
plt.ylim(0,80)
plt.xlabel('Age')
plt.ylabel('Frequency');
```

## Barplot 

```python
df['Pclass'].plot.bar();
```
