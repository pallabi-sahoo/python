```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('titanic_train.csv')
df.head()
```

## Piechart 

```python
dff=df['Pclass'].value_counts().sort_index()
dff
```

```python
colors=['#191970', '#001CF0', '#0038E2']
explode=(0,0.01,0.02)
```

```python
dff.plot(kind='pie',colors=colors, explode=explode, autopct='%1.1f%%', fontsize=12, figsize=(10, 10));
```

## Line plot

```python
df['Fare'].plot(figsize=(12,5),color='Red')
plt.xlim(0,500)
plt.ylim(0,700)
plt.xlabel('index')
plt.ylabel('Fare');
```

## Scatter Plot 

```python
df.plot.scatter('Age','Fare',figsize=(12,5))
```

## Box Plot 

```python
df['Age'].plot.box(figsize=(8,5));
```

```python
df.boxplot(figsize=(12,5));
```

## Histogram 

```python
df['Age'].hist(bins=30,figsize=(12,5));
plt.xlim(0,90)
plt.ylim(0,80)
plt.xlabel('Age')
plt.ylabel('Frequency');
```

## Barplot 

```python
df_group=df.groupby('Age')['Survived'].mean()
df_group
```

```python
df_group[:10].plot.bar(figsize=(12,5));
```

```python
df_group[:10].plot.barh(figsize=(12,8));
```

```python
ax=df[['Pclass','Age']][:5].plot.bar(stacked=True)
ax.set_xlabel("Index")
ax.set_ylabel("Count");
```

## Multiple Plots 

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
df['Age'].plot(ax = axes[0][0], style='.', color='red') # top left
df['Fare'].plot(ax = axes[0][1], style='.', color='blue') # top right
df['Fare'].plot.hist(bins=30, ax = axes[1][0], color='black') # bottom left
df['Age'].plot.hist(bins=20, ax = axes[1][1], color='gray'); #bottom right

axes[0][0].set_xlabel('index')
axes[0][1].set_xlabel('index')
axes[1][0].set_xlabel('Fare')
axes[1][1].set_xlabel('Age')
axes[0][0].set_ylabel('Age')
axes[0][1].set_ylabel('Fare')
axes[0][0].set_ylim(0, 100)
axes[0][1].set_ylim(0, 240)
axes[1][0].set_xlim(0, 200)
axes[1][1].set_xlim(0, 90);
```

```python
##Another method for Multiple plots
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
plt.subplot(2, 2, 1)
df['Age'].plot(style='.', color='red')
plt.subplot(2, 2, 2)
df['Fare'].plot(style='.', color='blue') # bottom right
plt.subplot(2, 2, 3)
df['Fare'].plot.hist(bins=30, color='black') # top right
plt.subplot(2, 2, 4)
df['Age'].plot.hist(bins=20, color='gray') # bottom left
plt.tight_layout()
plt.show();
```
